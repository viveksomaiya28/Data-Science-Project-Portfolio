```python
import pandas as pd
data = pd.read_csv('data/Telco-Customer-Churn.csv')

print("First 5 rows of the dataset:")
display(data.head())

print("\nDataset shape (rows, columns):", data.shape)

print("\nColumn names:")
print(data.columns.tolist())

print("\nData types and non-null counts:")
print(data.info())

print("\nMissing values per column:")
print(data.isnull().sum())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')  
%matplotlib inline

# Data Cleaning
print("Checking TotalCharges for non-numeric values:")
print(data['TotalCharges'].str.isspace().sum(), "rows with whitespace")
print(data['TotalCharges'].str.len().eq(0).sum(), "rows with empty strings")

# replaces empty strings or whitespace with NaN
data['TotalCharges'] = data['TotalCharges'].replace(r'^\s*$', np.nan, regex=True)

# Convert TotalCharges to float
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

# Check for NaN values after conversion
print("\nMissing values in TotalCharges after conversion:")
print(data['TotalCharges'].isnull().sum())

# Investigate NaN values (likely tenure=0)
nan_rows = data[data['TotalCharges'].isnull()]
print("\nRows with missing TotalCharges:")
print(nan_rows[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']])

# Replace NaN in TotalCharges with 0 (since tenure=0 implies no charges)
data['TotalCharges'] = data['TotalCharges'].fillna(0)

# Verify TotalCharges is now float64
print("\nData types after fixing TotalCharges:")
print(data[['TotalCharges']].dtypes)


print ("\nNumber of duplicate rows:", data.duplicated().sum())
print("Number of duplicate customerIDs:", data['customerID'].duplicated().sum())

# checks categorical variables
categorical_cols = data.select_dtypes(include='object').columns
print("\nUnique values in categorical columns:")
for col in categorical_cols:
    if col != 'customerID':  # Skip customerID
        print(f"{col}: {data[col].unique()}")


numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
print("\nSummary statistics for numerical features:")
print(data[numerical_cols].describe())

#analyzes target variable (Churn)
print("\nChurn distribution (counts):")
print(data['Churn'].value_counts())
print("\nChurn distribution (percentages):")
print(data['Churn'].value_counts(normalize=True) * 100)

plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=data, palette='Set2')
plt.title('Churn Distribution', fontsize=14)
plt.xlabel('Churn', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('visualisations/churn_distribution.png', dpi=300, bbox_inches='tight')  # High-quality save
plt.show()

data.to_csv('data/telco_churn_cleaned.csv', index=False)
print("\nCleaned dataset saved as 'data/telco_churn_cleaned.csv'")
```
