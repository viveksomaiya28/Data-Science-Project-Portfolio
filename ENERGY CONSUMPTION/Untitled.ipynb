```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("train_energy_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
for col in ["Building Type", "Day of Week"]:
    print(df[col].unique())
```

```python
df_encoded = df.copy()

df_encoded['Day of Week'] = df_encoded['Day of Week'].map({
    'Weekday': 0,
    'Weekend': 1
})

df_encoded['Building Type'] = df_encoded['Building Type'].map({
    'Residential': 0,
    'Commercial': 1,
    'Industrial': 2
})
corr_matrix = df_encoded.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Energy Consumption Features")
plt.show()
```
