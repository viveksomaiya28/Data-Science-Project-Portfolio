```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Exam_Score_Prediction.csv")
df.head()
```

```python
df = df.drop("student_id", axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
sns.pairplot(data=df, diag_kind = "kde", kind = "scatter", hue = "exam_difficulty", palette='husl')
plt.show()
```

```python
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = "coolwarm")
plt.show()
```

```python
df.columns
```

```python
internet_access = df["internet_access"].value_counts()
print(internet_access)
```

```python
exam_difficulties = df["exam_difficulty"].value_counts()
print(exam_difficulties)
```

```python
df.describe()
```

```python
sns.histplot(df['exam_score'], kde=True)
plt.title('Distribution of Exam Scores')
plt.show()
```

```python
sns.histplot(df['study_hours'], kde=True)
plt.title('Distribution of Study Hours')
plt.show()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='course', data=df)
plt.title('Course-wise Student Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(
    x='gender',
    y='exam_score',
    data=df,
    errorbar=('ci', 95)
)
plt.title('Average Exam Score by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='study_method', y='exam_score', data=df)
plt.title('Exam Score by Study Method')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.scatterplot(x='study_hours', y='exam_score', data=df)
plt.title('Study Hours vs Exam Score')
plt.show()
```

```python
sns.lmplot(x='study_hours', y='exam_score', data=df)
plt.title('Study Hours vs Exam Score (Trend)')
plt.show()
```

```python
sns.scatterplot(x='sleep_hours', y='exam_score', data=df)
plt.title('Sleep Hours vs Exam Score')
plt.show()
```

```python
sns.barplot(x='study_method', y='exam_score', data=df, ci=95)
plt.title('Average Exam Score by Study Method')
plt.xticks(rotation=30)
plt.show()
```

```python
sns.stripplot(x='gender', y='exam_score', data=df, jitter=True)
plt.title('Individual Exam Scores by Gender')
plt.show()
```

```python
sns.pointplot(
    x='class_attendance',
    y='exam_score',
    data=df,
    errorbar=('ci', 95)
)

plt.title('Exam Score Trend Across Attendance Levels')
plt.show()
```

```python
sns.catplot(
    x='exam_difficulty',
    y='exam_score',
    data=df,
    kind='bar',
    ci=95
)
plt.title('Average Score by Exam Difficulty')
plt.show()

```

```python
sns.violinplot(x='study_method', y='exam_score', data=df, inner=None)
sns.stripplot(x='study_method', y='exam_score', data=df, color='black', alpha=0.4)
plt.xticks(rotation=30)
plt.title('Exam Score Distribution by Study Method')
plt.show()
```

```python
sns.pointplot(
    x='class_attendance',
    y='exam_score',
    data=df,
    errorbar = ('ci',95)
)
plt.title('Trend of Exam Score Across Attendance Levels')
plt.show()
```

```python
mean_scores = df.groupby('exam_difficulty')['exam_score'].mean()

mean_scores.plot(marker='o')
plt.title('Average Exam Score vs Exam Difficulty')
plt.xlabel('Exam Difficulty Level')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
sns.barplot(
    x='sleep_quality',
    y='exam_score',
    data=df,
    order=sorted(df['sleep_quality'].unique()),
    ci=95
)
plt.title('Average Exam Score by Sleep Quality')
plt.show()
```

```python
sns.pointplot(
    x='facility_rating',
    y='exam_score',
    data=df,
    ci=95
)

sns.stripplot(
    x='facility_rating',
    y='exam_score',
    data=df,
    color='black',
    alpha=0.4
)

plt.title('Exam Score vs Facility Rating')
plt.show()
```
